```python

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def print_bboxes(path, label_size=1):

    img = np.array(Image.open(path))
    plt.figure()
    fig, ax = plt.subplots(1)
    ax.imshow(img)

    label_path = path.replace("images/", "labels/").replace(".png", ".txt").replace(".jpg", ".txt")

    boxes = np.loadtxt(label_path).reshape(-1, 4 + label_size)

    print(img.shape)

    for b in boxes:
        x,y,w,h = b[label_size:]


        x1 = (x - (w/2)) * img.shape[1]
        y1 = (y - (h/2)) * img.shape[0]

        w = w * img.shape[1]
        h = h * img.shape[0]

        bbox = patches.Rectangle(
            (x1, y1),
            w,
            h,
            linewidth=2,
            edgecolor=[0.8,0.4,0.2],
            facecolor="none",
        )

        ax.add_patch(bbox)

plt.rcParams["figure.figsize"] = (10,10)

print_bboxes("/mnt/c/Users/nickb/Documents/PyTorch-YOLOv3/data/urbanrecall/images/UrbanRecallTrain_0006030.png", 3)

print_bboxes("/mnt/c/Users/nickb/Documents/PyTorch-YOLOv3/data/coco/images/train2014/COCO_train2014_000000000030.jpg", 1)
```
